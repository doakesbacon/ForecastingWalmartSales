```python
#import dependencies
import pandas as pd
```

```python
#read in data
features = pd.read_csv("Resources/features.csv")
stores = pd.read_csv("Resources/stores.csv")
train = pd.read_csv("Resources/train.csv")
```

```python
features_stores_merge = pd.merge(features,stores,how='left')
features_stores_merge
```

```python
all_merge = pd.merge(features_stores_merge,train,how='right')
all_merge
```

```python
all_merge.head(50)
```

```python
holiday_sales = all_merge.loc[(all_merge["IsHoliday"] == True)]
holiday_sales.reset_index(inplace=True,drop=True)
holiday_sales
```

```python
holiday_sales.info()
```

```python
holiday_sales["Date"] = pd.to_datetime(holiday_sales["Date"])
holiday_sales["Month"] = holiday_sales["Date"].dt.month
```

```python
holiday_sales.info()
```

```python
holiday_sales
```

```python
holiday_map = {2: "Super Bowl", 9: "Labor Day", 11: "Thanksgiving", 12: "Christmas"}
holiday_sales["Holiday"] = holiday_sales["Month"].map(holiday_map)
holiday_sales
```

```python
others = holiday_sales.loc[(holiday_sales["Holiday"] == "Other")]
others
```
